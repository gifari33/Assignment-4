```python


import pandas as pd
import nltk
from nltk.tokenize import word_tokenize

# Bersihkan dan unduh ulang paket NLTK
nltk.data.path.clear()
nltk.download('punkt', force=True)

# Path file input dan output
input_path = "/content/drive/MyDrive/analisa sentimen git/Gojek🏍 App Reviews.csv"
output_path = "/content/drive/MyDrive/analisa sentimen git/Gojek_Processed.csv"

# Membaca file CSV
df = pd.read_csv(input_path)

# Hapus spasi berlebihan pada semua kolom string
df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
# Tokenisasi kolom 'content'
df['tokens'] = df['content'].apply(lambda x: word_tokenize(x) if isinstance(x, str) else [])

# Simpan ke file CSV baru
df.to_csv(output_path, index=False)

print(f"Proses selesai. File telah disimpan di: {output_path}")
```
